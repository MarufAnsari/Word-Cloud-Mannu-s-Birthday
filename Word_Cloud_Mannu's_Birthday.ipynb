#**Word Cloud : Mannu's Birthday**
“Word clouds (also known as text clouds or tag clouds) work in a simple way: the more a specific word appears in a source of textual data (such as a speech, blog post, or database), the bigger and bolder it appears in the word cloud.”


First of all, let’s import all the primary libraries first.

```
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
from PIL import Image # converting images into arrays
import matplotlib.pyplot as plt # for visualizing the data
from google.colab import files # for uploading files
from io import BytesIO
```

Now let’s import the package and it's set of stopwords.

```
from wordcloud import WordCloud, STOPWORDS
```

Word clouds are commonly used to perform high-level analysis and visualization of text data. Let’s try to analyze a short self synthesid text file describing Mannu's life.
Let’s go ahead and open the uploaded text file.

```
mannu_novel = open('text_m.txt', 'r').read()
print ('File downloaded and saved!')
```

Next, let’s use the stopwords that we imported from ‘word_cloud’. We use the function ‘set’ to remove any redundant stopwords and Create a word cloud object and generate a word cloud.

```
stopwords = set(STOPWORDS)
# instantiate a word cloud object
mannu_wc = WordCloud(
    background_color='black',
    contour_width=2,
    max_font_size=15,
    min_font_size=2,
    contour_color='white',
    stopwords=stopwords
)
# generate the word cloud
mannu_wc.generate(mannu_novel)
```

Awesome! Now that the word cloud is created, let’s visualize it.

```
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_figwidth(15) # set width
fig.set_figheight(20) # set height

# display the word cloud
plt.imshow(mannu_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Much better! However, said isn’t really an informative word. So let’s add it to our stopwords and re-generate the cloud.

```
stopwords.add('Rathi') # add the words said to stopwords
# re-generate the word cloud
mannu_wc.generate(mannu_novel)
# display the cloud
fig = plt.figure()
fig.set_figwidth(15) # set width
fig.set_figheight(20) # set height
plt.imshow(mannu_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Excellent! This looks really interesting! Another cool thing you can implement with the word_cloud package is superimposing the words onto a mask of any shape. Let's use a mask of mannu. I have already created the mask , so let's go ahead and use the above uploaded mask.

Let’s take a look at how the mask looks like.

```
mannu_im = Image.open(BytesIO(uploaded['PicsArt_11-07-03.29.20 (2).jpg']))
mannu_mask = np.array(mannu_im)
fig = plt.figure()
fig.set_figwidth(8) # set width
fig.set_figheight(12) # set height
plt.imshow(mannu_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Shaping the word cloud according to the mask is straightforward using `word_cloud` package.

(Note : you can try run this part many times for better wordcloud)

```
# instantiate a word cloud object
mannu_wc = WordCloud(background_color='black', mask=mannu_mask, stopwords=stopwords)
# generate the word cloud
mannu_wc.generate(mannu_novel)
# display the word cloud
fig = plt.figure()
fig.set_figwidth(8) # set width
fig.set_figheight(12) # set height
plt.imshow(mannu_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
